Design

To begin, I want to talk about what the simplest pocket area metric would be, and then go into some of its failings, and explain why this system is impractical, and then we will propose an improved system. Suppose you were to define the pocket area with a web of linearly connected lines between the rushing defensive players, on a typical 4-3 defensive scheme standard rush. Now, what makes this system of measuring pocket area ineffective? For starters, the QB is not considered to be in the pocket! This is despite the fact that the QB is clearly contained within the standard expectations of the pocket. By using a linear web between the rushers, you create this kind of opportunity for the system to miss whether the quarterback is actually within the pocket.

If you look at the next two examples and compare them, you will notice another failure of this system.




Although these pockets are valued the same, the second leaves less space for the Quarterback to move, as the Quarterback cannot move as easily directly to the left, otherwise, they will run into R2 (the second rusher). So, a more evenly spaced set of rushers better contains the quarterback and limits their space to adjust, thus making the area of the pocket smaller. The linear web of the rushers fails to acknowledge this and instead values the pocket areas as the same.

Now, we will propose a superior system. What if, instead of this linear web system between rushers, we used a half-circle system alongside the linear web, combining the linear web with these half-circles to assess the total area? This way, we can penalize rushers for leaving gaps. Say, for example, we focus on only the left side of the pocket in the above examples. Then, we get the following different pocket images.



Now that we use half circles to assess the pocket area, the total area size assessed in the second example is more significant than that of the first. This is intuitive, as there is more room for the QB to move into, due to the large gap between rushers.

To just give a quick example of how this area changes as the rushers move, let’s say in the first example, each rusher is 4 yards away from each other, and in the second, the first two are 2 yards away, with the second to the third having a distance of six yards, so in each instance, they are each 8 yards away total from each other within the same space. In the first, the half circles add in a total of 2(π2^(2)) of square yards of space to the pocket area, so 8π. In the second, they add π3^(2) + π1^(2) of total space in square yards, or 10π. So, the difference in total pocket area is 2π square yards, roughly 6 square yards of space added when there is a larger gap between one set of rushers in this instance.

Areas of Further Research

Although I believe this system, combining the area of the linear shape between rushers as well as half-circles between them, is useful and a great basis for assessing the available space in the pocket, or a QB’s presence in the pocket, I think it is certainly not without the need for further considerations and improvements.
    
Firstly, a key question is what rushers get included in the web that is used to measure space. What if there is a rusher who has fallen behind the others, and is not reasonably having any real effect on pocket size? This question will need further research, but my simplest, imperfect answer available at this time is that the combination of rushers used could be determined by the minimum pocket area assessed, i.e the combination of rushers that creates the smallest pocket area which includes the QB within it, with a minimum of two rushers in order to form the half circles between them on either side, at which point the QB is held within a circle if there is only two. This way, you can exclude rushers whose presence only damages the pocket area assessed. 

Another way of selecting participating rushers could be to only select rushers who are within the linear web, or half-circles, of other combinations of rushers, to eliminate outliers. This way, you create the pocket spaces but eliminate those that are not reasonably considered a part of any reasonable pocket that does not include them, so they could only have a negative impact on the space considered. Actually, the half-circle and linear web combined system for assessing pocket area allows for both of these styles of picking rushers, another strength of the system alongside those previously mentioned. If the system was purely a linear web, the pocket could only be assessed with at least 3 rushers, and you could not use the half circles as a way of noting those outside of all of them, as well as outside of the linear web, as rushing outliers. So, once again, we see the strengths of our combined system.
    
A second area of further research is how you consider rushers who are blocked or unblocked. One way in which I think this could potentially be accounted for is by allowing rushers who are free of blockers to invert their half circles, cutting into the linear web with their connections to others, as they are penetrating the space, and so they have a higher range of motion and so can more quickly adjust to eliminate range of motion for the QB.
    
Finally, a final area of further contemplation is the failure of this system to value running opportunities/gaps forward as compared to backward, but I did not feel as though it was fair to make a difference in valuation between these gaps available to QBs, as it depends on the Quarterback. There is little difference in valuation for a Quarterback who is not mobile and unlikely to scramble between forwards gaps in space to move into then throw or sideways/backward gaps to move into then throw. For a Quarterback who is extremely mobile and is a threat to run beyond the line of scrimmage with the ball, there is a significant difference in valuation. So, I felt as though setting a uniform difference in valuation between forward gaps in rushers and backward gaps in rushers would not be an effective assessment of the circumstances in any given situation. Instead, I feel as though this pocket area assessment should be used alongside another metric that assesses potential QB running risk by gap between rushers, wherein this other metric considers the traits of the given QB.

Summary

Again, I apologize, as I acknowledge the simplicity of my system amongst competing bodies of research that I am sure will be much more complicated. That being said, I think the simplicity of my system should not undermine its intuition and significant utility. I think assessing pocket area as a summed combination of a linear web between participating rushers and half-circles added to the exterior between them is a superior way of assessing the size of a pocket that accurately assesses the ability of a QB to move into space, and accurately assesses whether they are part of the pocket at all. It also is a system that is not overly complicated, and so is easier for traditional coaches to understand and utilize, as compared to more rigorous statistical work.

Obviously, as mentioned previously, there is much further work to be done. That being said, I think the suggestions given on how to improve this system on further study, as well as the system itself, provide a great basis on which to assess the pocket, and I think these ideas will provide value to both coaches who seek to protect their QBs, as well as coaches who seek to make the other team’s QBs as uncomfortable as possible.
